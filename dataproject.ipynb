```python
import folium
from folium.plugins import HeatMap
```

```python
sample_df=df.sample(int(0.01*len(df)))
lat_lng_pair=list(zip(list(sample_df['Start_Lat']),list(sample_df['Start_Lng'])))
```

```python
map=folium.Map()
HeatMap(lat_lng_pair).add_to(map)
map
```
